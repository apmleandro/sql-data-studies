```python
cursor.execute('''SELECT
                  g.name,
                  COUNT (DISTINCT ab.title)
                  FROM genres g
                  INNER JOIN tracks t
                  ON g.id = t.genre_id
                  INNER JOIN albums ab
                  ON t.album_id = ab.id
                  GROUP BY g.name
                  ORDER BY
                  COUNT (DISTINCT ab.title) DESC
                  LIMIT 1
                  ;''').fetchall()[0]
```

```python
cursor.execute(''' SELECT 
                c.first_name || " " || c.last_name,
                COUNT(DISTINCT i.id),
                SUM(il.unit_price)
                FROM customers c 
                JOIN invoices i
                ON c.id = i.customer_id
                JOIN invoice_lines il
                ON i.id = il.invoice_id
                GROUP BY c.first_name || " " || c.last_name
                ORDER BY SUM(il.unit_price) DESC
                LIMIT 5
                ''').fetchall()
```
